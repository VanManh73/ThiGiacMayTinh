```python
# Vẽ đồ thị cơ bản
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]
plt.plot(x, y, color='red') # Màu sắc
plt.title('Ví dụ về biểu đồ') # Title
plt.show() # Hiển thị
```

```python
# Tắt hệ trục tọa độ
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.plot(x, y, color='red')  # Thay đổi màu thành đỏ
plt.title('Biểu đồ ví dụ với màu đỏ')  # Thêm tiêu đề
# Tắt hệ trục tọa độ
plt.axis('off')  # Tắt hệ trục tọa độ
plt.show()
```
